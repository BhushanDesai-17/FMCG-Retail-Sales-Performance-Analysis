```python
import pandas as pd
```

```python
df = pd.read_csv('FMCG Sales Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df
```

```python
# Convert Invoice_Date Column

df['Invoice_Date'] = pd.to_datetime(df['Invoice_Date'], 
                                    format = '%d-%m-%Y %H:%M')
df
```

```python
# Additional Column

df['Month'] = df['Invoice_Date'].dt.month_name()
df['Day'] = df['Invoice_Date'].dt.day_name()
df
```

```python
# Handling Missing Customer_Age Column

df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].median())
df
```

```python
# Handling Customer_Gender Column

df['Customer_Gender'] = df['Customer_Gender'].fillna('Unknown')
df
```

```python
df.drop_duplicates()
df
```

```python
df['Customer_Gender'] = df['Customer_Gender'].replace('M', 'Male')
df['Customer_Gender'] = df['Customer_Gender'].replace('F', 'Female')
df['Customer_Gender'] = df['Customer_Gender'].replace('O', 'Other')
df
```

```python
# Data Validation
# Revenue Validation

df['Calculated_Revenue'] = df['Units'] * df['Selling_Price']
df
```

```python
# Cost Validation

df['Calculated_Cost'] = df['Units'] * df['Cost_Price']
df
```

```python
# MArgin Validation

df['Calculated_Margin'] = df['Calculated_Revenue'] - df['Calculated_Cost']
df
```

```python
# Saving the Cleaned FMCG DataSet

df.to_csv('FMCG Cleaned Dataset.csv', index = False)
df
```
